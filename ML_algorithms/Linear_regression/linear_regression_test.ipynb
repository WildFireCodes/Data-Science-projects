```python
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

<h1> 1. Regression from scikit-learn </h1>

```python
X, y = make_regression(n_samples = 800, n_features = 1, random_state = 42, noise = 7)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

Linear regression model error from scikit learn

```python
rmse_scikit = np.sqrt(np.mean((y_pred - y_test) ** 2))
rmse_scikit
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X, y)
plt.plot(X, regressor.predict(X), color = 'red')
```

<h1> 2. Regression from scratch </h1>

```python
from linear_regression import LinearRegression as LR
```

```python
regressor2 = LR(n_iterations=10000)
c = regressor2.fit(X_train, y_train)
y_pred2 = regressor2.predict(X_test)
```

```python
plt.figure(figsize=(8,8))
plt.scatter(X, y)
plt.plot(X, regressor2.predict(X), color = 'red')
```

Own linear regression model RMSE

```python
rmse_own = np.sqrt(np.mean((y_pred2 - y_test) ** 2))
rmse_own
```

```python
print(f'Scikit-learn: {rmse_scikit} vs own model: {rmse_own}')
```

```python
f = plt.figure()
f, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,8))
ax1.scatter(X, y)
ax1.plot(X, regressor2.predict(X), color = 'red')
ax1.set_title('Regression from scratch')

ax2.scatter(X, y)
ax2.plot(X, regressor.predict(X), color = 'red')
ax2.set_title('Regression from sklearn')
```
